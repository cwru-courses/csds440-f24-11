## Printing First Test in Voting Dataset

```python
import os
import sys
import numpy as np

# Add the 'src' directory to sys.path
cwd = os.getcwd()
project_root = os.path.abspath(os.path.join(cwd, '..'))  # Adjust as needed
src_path = os.path.join(project_root, 'src')
if src_path not in sys.path:
    sys.path.append(src_path)

# Import dtree function and other necessary components from dtree.py
from dtree import dtree, DecisionTree, evaluate_and_print_metrics,parse_c45

# Import util if needed
import util

# For data loading
from sting.data import parse_c45

# Set data path
data_path = os.path.join(project_root, '440data')  # Adjust as needed

# Verify the data path
print("Data path:", data_path)
```

```python
# Dataset name
dataset_name = 'voting'

# Construct the full path to the dataset
dataset_folder = os.path.join(data_path, dataset_name)

# Load the dataset using parse_c45
schema, X, y = parse_c45(dataset_name, data_path)

# Convert data types for consistency
X = X.astype(np.float32)
y = y.astype(np.int8)

print(f"Loaded {dataset_name} dataset with {X.shape[0]} instances and {X.shape[1]} features.")

# Initialize the decision tree with depth=1
decision_tree = DecisionTree(schema, max_depth=1, criterion='gain_ratio')

# Fit the decision tree to the entire dataset
decision_tree.fit(X, y)

# Get the first test (feature and threshold)
first_test = decision_tree.get_first_test()
print(f"\nFirst Test in Voting Dataset: {first_test}")
```
